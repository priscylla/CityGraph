```python
def get_inmet_hourly_data(station):
    
    station = str(station)
    
    #res06 = requests.get('https://apitempo.inmet.gov.br/estacao/2006-01-01/2006-12-31/' + station)
    res07 = requests.get('https://apitempo.inmet.gov.br/estacao/2007-01-01/2007-12-31/' + station)
    res08 = requests.get('https://apitempo.inmet.gov.br/estacao/2008-01-01/2008-12-31/' + station)
    res09 = requests.get('https://apitempo.inmet.gov.br/estacao/2009-01-01/2009-12-31/' + station)
    res10 = requests.get('https://apitempo.inmet.gov.br/estacao/2010-01-01/2010-12-31/' + station)
    res11 = requests.get('https://apitempo.inmet.gov.br/estacao/2011-01-01/2011-12-31/' + station)
    res12 = requests.get('https://apitempo.inmet.gov.br/estacao/2012-01-01/2012-12-31/' + station)
    res13 = requests.get('https://apitempo.inmet.gov.br/estacao/2013-01-01/2013-12-31/' + station)
    res14 = requests.get('https://apitempo.inmet.gov.br/estacao/2014-01-01/2014-12-31/' + station)
    res15 = requests.get('https://apitempo.inmet.gov.br/estacao/2015-01-01/2015-12-31/' + station)
    res16 = requests.get('https://apitempo.inmet.gov.br/estacao/2016-01-01/2016-12-31/' + station)
    res17 = requests.get('https://apitempo.inmet.gov.br/estacao/2017-01-01/2017-12-31/' + station)

    responses = []
    #responses.append(pd.DataFrame(json.loads(res06.text)))
    responses.append(pd.DataFrame(json.loads(res07.text)))
    responses.append(pd.DataFrame(json.loads(res08.text)))
    responses.append(pd.DataFrame(json.loads(res09.text)))
    responses.append(pd.DataFrame(json.loads(res10.text)))
    responses.append(pd.DataFrame(json.loads(res11.text)))
    responses.append(pd.DataFrame(json.loads(res12.text)))
    responses.append(pd.DataFrame(json.loads(res13.text)))
    responses.append(pd.DataFrame(json.loads(res14.text)))
    responses.append(pd.DataFrame(json.loads(res15.text)))
    responses.append(pd.DataFrame(json.loads(res16.text)))
    responses.append(pd.DataFrame(json.loads(res17.text)))
    
    data_station = pd.concat(responses)
    return data_station
```
